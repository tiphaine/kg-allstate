```python
% matplotlib inline

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [15.0, 8.0]
```

```python
train_file = 'train.csv'
test_file = 'test.csv'
```

```python
train_df = pd.read_csv(train_file)
```

```python
target_df = train_df[['loss']]
```

```python
categorical_data = []
continuous_data = []

for column_name in train_df.columns.tolist():
    if column_name.startswith('cat'):
        categorical_data.append(column_name)
    elif column_name.startswith('cont'):
        continuous_data.append(column_name)
        
```

```python
cont_train_df = train_df[continuous_data]
cat_train_df = train_df[categorical_data]
```
